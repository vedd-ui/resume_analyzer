```python
#Import libraries
import pandas as pd
import re
```

```python
#Enter your resume description
resume = """Data Science and Machine Learning enthusiast with a strong foundation in Python, SQL, and data analysis. 
Skilled in working with structured data, building efficient solutions, and delivering accurate results with attention to detail. 
Experienced in handling data processing tasks, database management, and developing practical projects. 
Passionate about solving real-world problems using data-driven approaches and continuously improving technical skills.
Interested in Machine Learning, NLP, and data-driven applications"""
resume=resume.lower()
```

```python
#Skills Required
skills_list = {"python": ["python"],"machine learning": ["machine learning", "ml"],"db":["database","databases","db"],
               "sql": ["sql", "structured query language"],"nlp":["nlp","natural language processing"],"mysql":["mysql"],"mongodb":["mongodb"],
               "deep learning":["dl","deep learning"],
              "data analysis":["data analysis","data analytics"],
              "excel":["ms excel","excel","spreadsheet","spreadsheets","microsoft excel"],
              "tableau":["tableau","dashboard"],
              "power bi":["power bi","dashboard"],
              "data visualization":["data visualization"]}
```

```python
resume = re.sub(r'[^a-zA-Z ]', '', resume)
```

```python
#Skill Count
skill_count = {}
for skill, variations in skills_list.items():
    count = 0
    for i in variations:
        count += resume.count(i)
    if count > 0:
        skill_count[skill] = count
```

```python
#Entered Job Description
job_desc = """We are looking fot a Data Scientist with strong knowledge of Python, SQL, and ML.
Knowledge of DL is a plus. Experience with Excel, and dashboards is preferred."""
```

```python
job_desc=job_desc.lower()
job_desc = re.sub(r'[^a-zA-Z ]', '', job_desc)
```

```python
#Skills that match with the job description
job_skills = []
for skill, variations in skills_list.items():
    for var in variations:
        if re.search(r'\b'+var+r'\b',job_desc):
            job_skills.append(skill)
            break
print(job_skills)
```

```python
#Skills that match with both resume & job description
match_skill=[]
for i in job_skills:
    if i in skill_count:
        match_skill.append(i)
match_skill
```

```python
#Calculating match percentage
if len(job_skills)>0:
    match_percent=(len(match_skill)/len(job_skills))*100
else: match_percent=0
if match_percent > 80: print("Excellent match")
elif match_percent > 50: print("Moderate match")
else: print("Low match")
```

```python
#Missing/Required Skills
missin_skills=[]
for i in job_skills:
    if i not in match_skill:
        missin_skills.append(i)
missin_skills
```
